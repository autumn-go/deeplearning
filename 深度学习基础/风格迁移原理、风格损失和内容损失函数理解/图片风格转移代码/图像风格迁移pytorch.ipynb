```python
#,
style = '/Users/xieshenqiu/Downloads/1200px-Van_Gogh_-_Starry_Night_-_Google_Art_Project.jpg'

#
content = '/Users/xieshenqiu/Downloads/IMG_20181122_232233.jpg'

#
style_weight=1000

#
content_weight=1

#
imsize = 128

loader = transforms.Compose([
    transforms.Resize(imsize),  # 
    transforms.ToTensor()])  # tensor

#
def image_loader(image_name):
    image = Image.open(image_name)
    image = loader(image).clone().detach().requires_grad_(True)
    # batch
    image = image.unsqueeze(0)
    return image

#
style_img = image_loader(style).type(dtype)
content_img = image_loader(content).type(dtype)

assert style_img.size() == content_img.size(), \
    ""
# 
def imshow(tensor, title=None):
    image = tensor.clone().cpu()  # Tensor
    image = image.view(3, imsize, imsize)  # batch
    image = unloader(image)
    plt.imshow(image)
    if title is not None:
        plt.title(title)
    plt.pause(0.001) # 

#
unloader = transforms.ToPILImage()  # PILPython Imaging Library 
plt.ion()

plt.figure()
imshow(style_img.data, title='Style Image')

plt.figure()
imshow(content_img.data, title='Content Image')
```

```python
cnn = models.vgg19(pretrained=True).features
if use_cuda:
    cnn = cnn.cuda()
```

```python
#
class ContentLoss(nn.Module):

    def __init__(self, target, weight):
        super(ContentLoss, self).__init__()
        # target
        self.target = target.detach() * weight
        self.weight = weight
        self.criterion = nn.MSELoss()

    def forward(self, input):
        # input
        # target
        self.loss = self.criterion(input * self.weight, self.target)
        self.output = input
        return self.output

    def backward(self, retain_graph=True):
        # 
        self.loss.backward(retain_graph=retain_graph)
        return self.loss

class StyleLoss(nn.Module):
# 
    def __init__(self, target, weight):
        super(StyleLoss, self).__init__()
        self.target = target.detach() * weight
        self.weight = weight
        #self.gram = GramMatrix()
        self.criterion = nn.MSELoss()

    def forward(self, input):
        # input
        self.output = input.clone()
        # gramtarget
        input = input.cuda() if use_cuda else input
        self_G = Gram(input)
        self_G.mul_(self.weight)
        # gramgram
        self.loss = self.criterion(self_G, self.target)
        return self.output

    def backward(self, retain_graph=True):
        # 
        self.loss.backward(retain_graph=retain_graph)
        return self.loss
    
    #Gram
def Gram(input):
    # gram
    a, b, c, d = input.size()  # a=batch size(=1)
    # b=
    # (c,d)= (N=c*d)

    features = input.view(a * b, c * d)  # 

    G = torch.mm(features, features.t())  # 

    # 
    return G.div(a * b * c * d)
```

```python
#  :
content_layers = ['conv_4'] #


style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']
# 12345


# 
content_losses = []
style_losses = []

model = nn.Sequential()  # 

# GPUGPU:
if use_cuda:
    model = model.cuda()


    
    
# vggmodel
# vgg
# model
i = 1
for layer in list(cnn):
    if isinstance(layer, nn.Conv2d):
        name = "conv_" + str(i)
        #model
        model.add_module(name, layer)

        if name in content_layers:
            # :
            target = model(content_img).clone() #featuretarget
            content_loss = ContentLoss(target, content_weight) #content_loss
            content_loss = content_loss if use_cuda else content_loss
            model.add_module("content_loss_" + str(i), content_loss) #content_loss
            content_losses.append(content_loss)

        if name in style_layers:
            # 
            target_feature = model(style_img).clone()
            target_feature = target_feature.cuda() if use_cuda else target_feature
            target_feature_gram = Gram(target_feature)
            style_loss = StyleLoss(target_feature_gram, style_weight)
            style_loss = style_loss.cuda() if use_cuda else style_loss
            model.add_module("style_loss_" + str(i), style_loss)
            style_losses.append(style_loss)
    if isinstance(layer, nn.ReLU):
        #
        name = "relu_" + str(i)
        model.add_module(name, layer)

        i += 1

    if isinstance(layer, nn.MaxPool2d):
        name = "pool_" + str(i)
        model.add_module(name, layer)  # ***
```
